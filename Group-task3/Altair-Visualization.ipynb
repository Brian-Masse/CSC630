```python
import pandas as pd
import altair as alt

xls = pd.ExcelFile(
    "/Users/brianmasse/Developer/Python/CSC630/CSC630/Group-task3/data/data_ord2.xlsx"
)

win_rates = pd.read_excel(xls, 'card win rate')
```

```python
pink="#f7b2d6"
dark_grey="#464c4d"
green="#b9d98f"
peach="#eef2d8"
dark_pink="#eb4da6"
med_pink="#f277b0"
```

```python
brush = alt.selection_interval(encodings=['x'])
opacity = alt.condition(brush, alt.value(1), alt.value(0.1))

overview = alt.Chart(win_rates).mark_bar().encode(
    alt.X('win%, rounded',
      axis=alt.Axis(title="Win% for any card", labelAngle=0) 
    ),
    alt.Y('number of win%', title="Number of Cards with win%"), 
    opacity=opacity
).add_selection(
    brush    
).properties(
    title="Number of win%",
    width=800, 
    height=50  
)

```

```python
chart = alt.Chart(win_rates)


detail = chart.mark_bar().encode(
    alt.X( "card ID:Q",
        title="cardID", axis=alt.Axis() ),
    alt.Y("uses:Q",
        title="Number of Card uses", axis=alt.Axis()),
    opacity=opacity,
    tooltip=[ "cardName", "uses", "win%, rounded"]
).properties(
    title="Card Uses",
    width=800
)

# alt.vconcat(overview, detail
# ).configure(
#     background=peach,
#     font="DINCondensed-Bold"
# ).configure_view (
#     strokeWidth = 0,
#     stroke='#ccc8a1',
#     fill='#f0ecce',
#     cornerRadius=10,
    
# ).configure_bar(
#     binSpacing=5.5,
#     cornerRadius=20,
#     cursor="help",
#     width=5,
#     color=green
# ).configure_axis(
#     labelFontSize=15,
#     titleFontSize=20,
#     titleColor=med_pink,
#     tickColor=pink,
#     domainColor=pink,
#     labelColor=pink,
#     grid=False,
# ).configure_title(
#     fontSize=30,
#     color=dark_pink 
# )
```

```python
background2="#c7c5b5"
highlight="#bf9b71"
title="#915f24"
bold_title="#693b04"
thin_title="#ad7a3d"

dominant_bar="#4F736D"
light_bar="#3B9686"
```

```python
base = alt.Chart(win_rates).encode(
    alt.X('elixer:Q', title="Card Elixer")
)

area = base.mark_bar(width=30, color=light_bar, opacity=0.7).encode(
    alt.Y('average(uses)', title="Number of Uses", axis=alt.Axis(
        titleColor=light_bar,
        laeblColor=light_bar
    ))
    
)

line = base.mark_bar(width=5,color=dominant_bar).encode(
    alt.Y('average(win%, rounded)', title="win %")
).properties(
    title="uses and win%, as a function of card elixer"
)

alt.layer(area, line).resolve_scale(
    y = 'independent'
).configure(
    background=background2,
    font="DINCondensed-Bold"
).configure_view(
    fill=highlight,
    fillOpacity=0.5,
    strokeWidth=0
).configure_title(
    fontSize=30,
    color=dominant_bar
).configure_axis(
    labelFontSize=13,
    titleFontSize=20,
    labelColor=thin_title,
    titleColor=title,
    domainColor=thin_title,
    tickColor=thin_title,
    gridColor=thin_title
).configure_bar(
    cornerRadius=8
)
```
